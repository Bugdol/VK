```python
import json
from html import unescape
import pandas as pd

data = []
with open('/content/drive/MyDrive/ranking_train.jsonl', 'r') as f:
    for line in f:
        json_data = json.loads(line.strip())
        post_text = unescape(json_data['text'])
        post_text = post_text.replace('\'', '').replace('\n', '').replace('\t', '')
        comments = json_data['comments']
        for comment in comments:
            comment_text = unescape(comment['text'])
            comment_text = comment_text.replace('\n', '').replace('\t', '').replace('http',' http').replace('[0]','').replace('[1]','').replace('[2]','').replace('[3]','').replace('[4]','').replace('[5]','')
            mist = '/\[]{}()&^%$#!"№;:<>_'
            for i in comment_text:
              if i in mist:
                comment_text = comment_text.replace(i, '')
              

            score = comment.get('score')
            data.append({'post_text': post_text, 'comment': comment_text, 'score': score})

df_train = pd.DataFrame(data)
```

```python
df_train.to_csv (r'/content/vk.csv.lnk', index = False)
```
